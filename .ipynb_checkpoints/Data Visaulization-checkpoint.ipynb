## Construct train and valid dataset

```python
train = nextDayPlayerEngagement[['playerId', 'target1', 'target2', 'target3', 'target4', 'date']].merge(
    my_dict['players'][['playerId', 'primaryPositionName', 'heightInches', 'weight']], on=['playerId'], how='left')
```

```python
train = train.rename(columns = {"primaryPositionName": "position", "heightInches": "height"})
```

```python
train = train.merge(train_dict['rosters'][['playerId', 'teamId', 'status', 'date']], on=['playerId', 'date'], how='left')
```

```python
score_cols = train_dict['playerBoxScores'].drop(['gameDate', 'gameTimeUTC', 'teamName', 'jerseyNum', 'positionCode',
                                                 'positionName', 'positionType', 'home', 'gamePk', 'teamId','index'],
                                                axis = 1).columns
```

```python
train = train.merge(train_dict['playerBoxScores'][score_cols], on=['playerId', 'date'], how='left')
```

```python
target_stats_2021 = pd.read_csv("data/player_target_stats_2021.csv")
target_stats_2020 = pd.read_csv("data/player_target_stats_2020.csv")
```

```python
target_stats_2020 = target_stats_2020.add_suffix('_2020')
target_stats_2020 = target_stats_2020.rename(columns = {"playerId_2020": "playerId"})
```

```python
train = train.merge(target_stats_2021, how='inner', left_on=["playerId"],right_on=["playerId"])
train = train.merge(target_stats_2020, how='inner', left_on=["playerId"],right_on=["playerId"])
```

```python
to_be_encoded = ['battingOrder','playerId', 'teamId', 'status', 'position']
```

```python
def label_encode(df, column):
    func = {c: i for i, c in enumerate(df[column].unique())}
    df[column] = df[column].map(func)

for i in to_be_encoded:
    label_encode(train, i)
```
